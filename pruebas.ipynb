# Prueba t-student

```python
# Seleccionando los filtro
alto_apoyo_intitucional = df[df["indice_apoyo_institucional"] > 1]["Have you ever sought treatment for a mental health issue from a mental health professional?"]
bajo_apoyo_intitucional = df[df["indice_apoyo_institucional"] < 1]["Have you ever sought treatment for a mental health issue from a mental health professional?"]

# Prueba t-student independiente y p-value
t_stat, p_value = stats.ttest_ind(alto_apoyo_intitucional, bajo_apoyo_intitucional)
t_stat, p_value
```

# Prueba chi cuadrado

```python
# Clasifico en dos grupos de nivel de apoyo institucional
df["apoyo"] = df["indice_apoyo_institucional"].apply(lambda x: "Alto" if x > 1 else "Bajo")

# Tabla de frecuencias cruzadas
tabla = pd.crosstab(df["indice_apoyo_institucional"], df["Have you ever sought treatment for a mental health issue from a mental health professional?"])

# Aplicar chi cuadrado
chi2_stats, p_value, dof, expected = stats.chi2_contingency(tabla)
chi2_stats, p_value
```
